## Resumen:

Este informe detalla las etapas clave del proyecto, con el objetivo principal de analizar los datos de la empresa para predecir la retención de clientes. El proceso se dividió en las siguientes fases:

## Importación de Bibliotecas y Datos

Se importaron las bibliotecas necesarias y se cargaron los dataframes de Internet, Contratos, Personal y Teléfono para su posterior análisis.

## Exploración de los Dataframes

Se exploraron los dataframes para comprender su contenido, los tipos de datos y datos ausentes.

## Limpieza de Datos

- En el dataframe 'Internet', se convirtieron las columnas 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', y 'StreamingMovies' de tipo object a bool.
- En el dataframe 'Contrato', se corrigieron las columnas 'BeginDate' y 'EndDate' convirtiéndolas a datetime, 'PaperlessBilling' a bool y 'TotalCharges' a float.
- En el dataframe 'Personal', las columnas 'Partner' y 'Dependents' se convirtieron a bool.
- En el dataframe 'Teléfono', la columna 'MultipleLines' se corrigió a tipo bool.

## Combinación de Dataframes

Los dataframes se combinaron mediante la columna 'customerID'.

## Selección de Características Relevantes

Se identificaron y seleccionaron las características más relevantes para el análisis.

## Codificación de Variables Categóricas

Se aplicó la codificación one-hot para las variables categóricas, en columnas como Type, Gender e InternetService.

Utilizamos one hot encoding para evitar que nuestro modelo malinterprete las categorías como si fueran números con un orden específico


## División de Datos

Los datos se dividieron en conjuntos de entrenamiento y prueba.

## Creación de Modelos

Se crearon varios modelos para predecir la retención de clientes.

Cada modelo tiene sus propias ventajas y desventajas, y la elección de cuál usar depende de diversos factores

Random Forest Modelo 1:

Procesa bien las características categóricas, y aguanta sobreajustes, además puede capturar relaciones no lineales en los datos. Es un modelo sólido para problemas de clasificación como la retención de clientes, donde se pueden tener características diversas y relaciones complejas.

Gradient Boosting Modelo 2:

Puede mejorar la precisión del modelo tiene la capacidad de tratar con relaciones no lineales y es adecuado para buscar un alto rendimiento predictivo

Support Vector Machine (Modelo 3):

Es efectivo en la separación de clases además de ser adecuado con datos no lineales y se busca una alta precisión en la clasificación.

Logistic Regression Modelo 4:

Es un modelo simple y fácil de interpretar nos da una comprensión clara de cómo cada característica afecta a la retención de clientes, y no se necesita un modelo muy complejo.


## Evaluación de la Exactitud

Se evaluó la exactitud de los modelos.

La evaluación de la exactitud de los modelos se llevó a cabo utilizando las métricas AUC-ROC y accuracy. Las cuales nos proporcionan información sobre el rendimiento de los modelos.

AUC-ROC:

Evalua la capacidad del modelo para clasificar correctamente las clases positivas y negativas.
Los criterios de evaluación para AUC-ROC que se proporcionaron son apropiados, esta metrica es útil para asignar una calificación a los modelos en función de su rendimiento.
Exactitud o Accuracy:

La exactitud mide la proporción  de predicciones correctas.

## Dificultades y Soluciones

Durante el proyecto, se encontraron dificultades, como la identificación y corrección de tipos de datos incorrectos en algunas columnas, las cuales se resolvieron mediante la conversión de tipos de datos y la limpieza de estos mismos.

## Pasos Clave para Resolver la Tarea

Los pasos clave para resolver la tarea incluyeron la limpieza y transformación de datos, la selección de características relevantes y la creación de modelos de predicción.

En resumen, el proyecto de análisis de datos de telecomunicaciones ha abordado con éxito la limpieza de datos y la preparación, pero se requiere más información sobre los modelos y su calidad para tomar decisiones finales.


## Resultados del Análisis de Datos

Durante el proceso de análisis de datos y la construcción de modelos, se obtuvieron las siguientes conclusiones:

Características Relevantes: Identificamos que las características relacionadas con el tipo de servicio de Internet, el tipo de contrato, los cargos mensuales y la antigüedad del cliente son factores claves para poder predecir la retención de clientes.

Desbalance de Clases: Observamos un desequilibrio en las clases de retención y no retención de clientes, donde la mayoría de los clientes se mantienen. Esto es importante tenerlo en cuenta al interpretar las métricas de rendimiento del modelo.

## Recomendaciones para el Negocio

Segmentación de Clientes: Utilizar los modelos de clasificación para segmentar a los clientes en función de su probabilidad de abandono. Esto permite identificar a los clientes de alto riesgo que requieren una atención especial y retención.

Programas de Retención Personalizados: Desarrollar programas de retención personalizados para los clientes identificados como de alto riesgo. Ofrecer incentivos y promociones a estos clientes para aumentar su lealtad.

Seguimiento Continuo: Implementar un sistema de seguimiento continuo para evaluar el rendimiento de los modelos y ajustar los parametros de clasificación según las necesidades que pueda tener el negocio.

Recopilación de Comentarios: Recopilar comentarios y retroalimentacion de los clientes que abandonan el servicio para entender las decisiones del cliente y mejorar la retención.
